## Plot the data

#### MAESTER

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))

#ax.flat[0].plot(dic_original_maester_embeddings[:100], label='centroid_maester')
ax.flat[0].plot(dic_baseline_no_embeddings[:100], label='dic baseline / no emb.', color='black', linestyle='dashed')

#ax.flat[0].plot(dic_vit_b_maester_embeddings[:100], label='centroid_vit_b')
ax.flat[0].plot(dic_original_maester_embeddings[:100], label='dic_original_maester_embeddings')

#ax.flat[0].plot(mma_dic_vit_b_maester_embeddings[:100], label='mma_vit_b')
ax.flat[0].plot(dic_mma_maester_embeddings[:100], label='dic_mma_maester_embeddings')


ax.flat[0].set_ylabel("Linking performance (LNK)")
ax.flat[0].set_xlabel("Training epochs")

ax.flat[0].legend()
ax.flat[0].set_title("DIC-C2DH-HeLa")

ax.flat[0].set_ylim(0.9, 1.0)
ax.flat[0].set_xlim(0, 20)

#ax.flat[1].plot(fluo_original_maester_embeddings[:100], label='centroid_maester')
ax.flat[1].plot(fluo_baseline_no_embeddings[:100], label='fluo baseline / no emb.', color='black', linestyle='dashed')

#ax.flat[1].plot(fluo_vit_b_maester_embeddings[:100], label='centroid_vit_b')
ax.flat[1].plot(fluo_original_maester_embeddings[:100], label='fluo_original_maester_embeddings')

#ax.flat[1].plot(mma_fluo_vit_b_maester_embeddings[:100], label='mma_vit_b')
ax.flat[1].plot(fluo_mma_maester_embeddings[:100], label='fluo_mma_maester_embeddings')

ax.flat[1].set_ylabel("Linking performance (LNK)")
ax.flat[1].set_xlabel("Training epochs")

ax.flat[1].legend()
ax.flat[1].set_title("Fluo-N2DL-HeLa")

ax.flat[1].set_ylim(0.97, 1.0)
ax.flat[1].set_xlim(0, 20)

plt.savefig("fluo_dic_draft.png")
```
