```python
import os
import pandas as pd
```

```python
blruns = list(set(["_".join(x.split("_")[:-1])for x in os.listdir("baseline")]))
embruns = list(set(["_".join(x.split("_")[:-1])for x in os.listdir("withembeddings")]))
```

```python
import os
import pandas as pd

IN_DIR = "baseline"
OUT_DIR = "without"

os.makedirs(OUT_DIR, exist_ok=True)

def numeric_suffix(fname: str) -> int:
    return int(fname.rsplit("_", 1)[1].replace(".csv", ""))

for rname in blruns:
    runs = [
        x for x in os.listdir(IN_DIR)
        if x.startswith(rname) and x.endswith(".csv")
    ]
    runs = sorted(runs, key=numeric_suffix)

    dfs = [pd.read_csv(os.path.join(IN_DIR, run)) for run in runs]
    combined = pd.concat(dfs, ignore_index=True)

    out_path = os.path.join(OUT_DIR, f"{rname}.csv")
    combined.to_csv(out_path, index=False)

    print(f"Wrote {out_path} ({len(combined)} rows from {len(runs)} files)")
```
