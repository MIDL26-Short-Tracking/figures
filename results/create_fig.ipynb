```python
combined = [list() for _ in range(300)]

for fname in os.listdir("data/dic/without"):
    df = pd.read_csv("data/dic/without/"+fname)

    l = df['LNK'].to_list()

    
    for idx, val in enumerate(l):
        if val > 0 and val < 999:
            combined[idx].append(val)



dic_baseline_no_embeddings = [sum(x) / len(x) if x else None for x in combined]

iloc = dic_baseline_no_embeddings.index(max(dic_baseline_no_embeddings))
print(combined[iloc])
mean = np.mean(combined[iloc])
variance = np.var(combined[iloc])
print(mean)
print(variance)
```
